### Garden Problem
What is the largest area that we can fence in a Garden using using <b> 100 meters </b> of fence.Define the Dimension of the garden as well.




```python
import pyomo.environ as pyo
from pyomo.environ import *
from pyomo.opt import SolverFactory 
```

```python
model= pyo.ConcreteModel()

model.x =pyo.Var(bounds=(0,None))
model.y =pyo.Var(bounds=(0,None))

x=model.x
y= model.y

model.obj =pyo.Objective(expr= x*y ,sense =maximize)
model.c1 = pyo.Constraint(expr =2*x+y <= 100)


opt= SolverFactory("ipopt")
opt.solve(model)


print("x is ",pyo.value(x))
print("y is",pyo.value(y))

print("Area is ",pyo.value(x)*pyo.value(y))
```
