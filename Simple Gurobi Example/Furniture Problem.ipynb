```python
# import gurobi library
from gurobipy import *
```

```python
# Creating the Model
f= Model("Furniture")
```

```python
#Creates Variables
x1= f.addVar(name="chairs")
x2= f.addVar(name="tables")
```

```python
#creating objective function
f.setObjective(45*x1 + 80*x2, GRB.MAXIMIZE)
```

```python
#setting constraint 
f.addConstr(5*x1 + 20*x2 <= 400,"mahagony")
f.addConstr(10*x1 + 15*x2 <=450 ,"Labor" )
```

```python
f.optimize()
```

```python
# Print the values of all variables
for v in f.getVars():
    print(f"{v.VarName} = {v.X}")
```

```python
print("The maximum value can be profited is", f.ObjVal)
```
